```python
import numpy as np
import os
from shapely.geometry import Polygon
```

```python
def parse_annotation_file(filepath):
    """Parse YOLO formatted annotation file to extract polygon masks."""
    masks = []
    with open(filepath, 'r') as file:
        for line in file:
            parts = line.strip().split()
            # Assuming YOLO format: class x1 y1 x2 y2 x3 y3 ... xn yn
            class_id = int(parts[0])
            points = np.array(parts[1:], dtype=float).reshape(-1, 2)
            masks.append((class_id, Polygon(points)))
    return masks

def compute_iou(poly1, poly2):
    """Compute the Intersection over Union (IoU) of two polygons."""
    if not poly1.is_valid:
        poly1 = poly1.buffer(0)
    if not poly2.is_valid:
        poly2 = poly2.buffer(0)
    inter_area = poly1.intersection(poly2).area
    union_area = poly1.union(poly2).area
    return inter_area / union_area

def compute_metrics(pred_masks, true_masks, iou_threshold=0.5):
    """Compute precision, recall, and F1-score based on IoU threshold."""
    TP = 0
    FP = 0
    FN = 0
    
    matched = []
    
    for true_class_id, true_mask in true_masks:
        matched_flag = False
        for i, (pred_class_id, pred_mask) in enumerate(pred_masks):
            if i in matched or pred_class_id != true_class_id:
                continue
            iou = compute_iou(pred_mask, true_mask)
            if iou >= iou_threshold:
                matched.append(i)
                TP += 1
                matched_flag = True
                break
        if not matched_flag:
            FN += 1
    
    FP = len(pred_masks) - len(matched)
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    return precision, recall, f1_score
```

```python
# Import necessary libraries
from ultralytics import YOLO
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
import torch

# Define paths
model_path = '/media/mzc/e4fcae9b-7400-4641-a800-034f249f8d14/mzc/yolo11RanjanSynthetic/aiIMAGEGENERATED/AutoAnnotation/AIsyntheticImagesAutoAnnotation/zeroshotYolo11/OutputYOLO11Segmentation/YOLO11m/train/weights/best.pt'
images_dir = '/media/mzc/e4fcae9b-7400-4641-a800-034f249f8d14/mzc/yolo11RanjanSynthetic/aiIMAGEGENERATED/AutoAnnotation/AIsyntheticImagesAutoAnnotation/zeroshotYolo11/OutputYOLO11Segmentation/LLMgeneratedApples.v1i.yolov11/valid/images'
output_dir = '/media/mzc/e4fcae9b-7400-4641-a800-034f249f8d14/mzc/yolo11RanjanSynthetic/aiIMAGEGENERATED/AutoAnnotation/AIsyntheticImagesAutoAnnotation/zeroshotYolo11/OutputYOLO11Segmentation/LLMgeneratedApples.v1i.yolov11/valid/Predicted'
blackwhite_dir = '/media/mzc/e4fcae9b-7400-4641-a800-034f249f8d14/mzc/yolo11RanjanSynthetic/aiIMAGEGENERATED/AutoAnnotation/AIsyntheticImagesAutoAnnotation/zeroshotYolo11/OutputYOLO11Segmentation/LLMgeneratedApples.v1i.yolov11/valid/blackwhite'
heatmap_dir = '/media/mzc/e4fcae9b-7400-4641-a800-034f249f8d14/mzc/yolo11RanjanSynthetic/aiIMAGEGENERATED/AutoAnnotation/AIsyntheticImagesAutoAnnotation/zeroshotYolo11/OutputYOLO11Segmentation/LLMgeneratedApples.v1i.yolov11/valid/Heatmap'

# Load the model
model = YOLO(model_path)

# Ensure the output directories exist
for dir_path in [output_dir, blackwhite_dir, heatmap_dir]:
    os.makedirs(dir_path, exist_ok=True)

def create_heatmap(masks, confidences, size=640):
    heatmap = np.zeros((size, size), dtype=np.float32)
    for mask, conf in zip(masks, confidences):
        if isinstance(mask, torch.Tensor):
            mask = mask.cpu().numpy()
        if isinstance(conf, torch.Tensor):
            conf = conf.cpu().numpy()
        mask = cv2.resize(mask, (size, size))
        heatmap += mask * conf
    return heatmap

# Process each image in the directory
for image_filename in os.listdir(images_dir):
    if not image_filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        continue

    image_path = os.path.join(images_dir, image_filename)
    
    # Load and segment the image
    img = cv2.imread(image_path)
    if img is None:
        print(f"Failed to load image: {image_path}")
        continue

    results = model(img, task='segment')

    # Draw segmentation masks on the image
    img_with_masks = results[0].plot()

    # Save the segmented image
    output_image_path = os.path.join(output_dir, f"segmented_{image_filename}")
    cv2.imwrite(output_image_path, img_with_masks)
    print(f"Saved segmented image: {output_image_path}")

    # Create black and white binary mask
    binary_mask = np.zeros((img.shape[0], img.shape[1]), dtype=np.uint8)
    masks = []
    confidences = []
    for r in results:
        if r.masks is not None:
            for seg, conf in zip(r.masks.data, r.boxes.conf):
                seg_mask = cv2.resize((seg.cpu().numpy() > 0.5).astype(np.uint8), (img.shape[1], img.shape[0]))
                binary_mask = cv2.bitwise_or(binary_mask, seg_mask * 255)
                masks.append(seg)
                confidences.append(conf)

    # Save the black and white binary mask
    blackwhite_image_path = os.path.join(blackwhite_dir, f"blackwhite_{image_filename}")
    cv2.imwrite(blackwhite_image_path, binary_mask)
    print(f"Saved black and white binary mask: {blackwhite_image_path}")

    # Create and save heatmap
    if masks:
        heatmap = create_heatmap(masks, confidences)
        plt.figure(figsize=(10, 10))
        plt.imshow(heatmap, cmap='jet')
        plt.colorbar(label='Confidence')
        plt.title('Segmentation Confidence Heatmap')
        plt.axis('off')
        heatmap_path = os.path.join(heatmap_dir, f"heatmap_{os.path.splitext(image_filename)[0]}.png")
        plt.savefig(heatmap_path, bbox_inches='tight', pad_inches=0.1)
        plt.close()
        print(f"Saved heatmap: {heatmap_path}")
    else:
        print(f"No masks found for image: {image_filename}")

print("All images have been processed and saved with segmentation masks, black and white binary masks, and heatmaps.")
```
